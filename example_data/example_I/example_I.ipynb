# Example I: Harmonically trapped 1D fermions with plain cutoff

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## $2\uparrow+2\downarrow$ benchmark 

Benchmark data digitized from Fig. 3 of [this paper](https://iopscience.iop.org/article/10.1088/1367-2630/17/11/115001).

```python
# Read results.
nb = 20 # Cutoff fixed.
n_part = (2,2) # Either (2,2) or (3,1)
e_offset = 4.0 # Noninteracting energy for comparison to literature.
data = pd.read_csv("exI_data_{:d}+{:d}.csv".format(*n_part))

with plt.style.context("seaborn"):
    fig, ax = plt.subplots()

    # Energy.
    for N in range(4):
        subset = data[data['N']==N+1].sort_values(by="coupling")
        line = ax.plot(subset["coupling"], subset["energy"], marker='o', ls='', label='' if N else 'this work ($n_b = 20$)', color='#B82200')

    # ----------------------------------------------------
    # ----------------------------------------------------
    # Benchmark section.

    # Energy, interpolated from digitized version.
    grin = pd.read_csv('grining_{:d}+{:d}.csv'.format(*n_part))
    for N in range(3):
        sub = grin[(grin['N']==N+1) & (grin['nb']==20)].sort_values(by='g')
        ax.plot(sub['g'], sub['energy']+e_offset, marker='', color='black', ls='-', lw=1, zorder=1, label='' if N else 'Grining et al., $n_b = 20$')

        sub = grin[(grin['N']==N+1) & (grin['nb']==0)].sort_values(by='g')
        ax.plot(sub['g'], sub['energy']+e_offset, marker='', color='gray', ls='-.', lw=0.5, zorder=0, label='' if N else 'Grining et al., $n_b\\to\\infty$')


    # ----------------------------------------------------
    # ----------------------------------------------------
    # Cosmetics & export.

    ax.axhline(8.0, ls='--', marker='', lw=0.25, color='black', zorder=0)

    ax.set_xlabel('$g$')
    ax.set_ylabel('$E$')
    ax.legend(loc='lower right')
    ax.set_ylim(3.0, 9.0)
```
